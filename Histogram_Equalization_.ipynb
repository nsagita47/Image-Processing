```
import matplotlib.pyplot as plt
```

```
x = (0,1,2,3,4,5,6,7)
y = (8,12,11,15,10,4,2,2)
```

```
plt.bar(x,y)
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')
plt.title('Histogram Citra Sebelum Histogram Equalization', fontsize=18)
plt.grid(color='gray', linestyle=':', linewidth=1.0)
```

```
# Jumlah pixel yang baru 
y_new = (0,1,1,1,0,1,1,3)
```

```
plt.bar(x,y_new)
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')
plt.title('Histogram Citra Setelah Histogram Equalization', fontsize=18)
plt.grid(color='gray', linestyle=':', linewidth=1.0)
```
